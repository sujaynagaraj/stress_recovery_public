# PROCESS SURVEY DATA

```python
EXTENSION = "/surveys_and_active_tasks/"
```

```python
once = [ 'participants.csv.gz',"demographics_survey.csv.gz","baseline_covid19_risk.csv.gz", "medical_history.csv.gz", "family_factors.csv.gz",
        "lifestyle_survey.csv.gz", "life_events_checklist_10item.csv.gz", "adverse_childhood_events_ace.csv.gz",
        "acceptabilitystudy_feedback.csv.gz", "ptsd_checklist_pclc.csv.gz", "personality_inventory_tipi.csv.gz"]


daily = ["daily_stress_measure.csv.gz", "cough_ema.csv.gz", "selfassessment_mannequin_sam.csv.gz",
        "ppe_and_exposure_scale.csv.gz"]

weekly = ["perceived_stress_scale_pss4.csv.gz", "promis_sleep_disturbance_sd.csv.gz",
         "promis_sleeprelated_impairment_sri.csv.gz", "ongoing_covid19_risk.csv.gz",
         "fatigue_severity_scale_fss9.csv.gz", "cough_vas.csv.gz"]

monthly = ["promis_global10.csv.gz", "promis_emotional_support.csv.gz", "rand_36item_short_form_health_survey.csv.gz",
          "survey_response_inclination_sri.csv.gz"]

other_week = ["patient_health_questionnaire_phq9.csv.gz", "generalized_anxiety_disorder_scale_gad7.csv.gz"]
```

```python
#DROPPING DUPLICATES KEEP FIRST
once_dfs = [pd.read_csv(DATAPATH+EXTENSION+NAME,compression='gzip').sort_values(by=['participant_id']).drop_duplicates(subset=["participant_id"], keep ="first") for NAME in once]
```

```python
once_merged = reduce(lambda  left,right: pd.merge(left,right,on=['participant_id', 'record_id'],
                                            how='outer'), once_dfs)
```

```python
df = once_merged.loc[:,~once_merged.columns.duplicated()]

#renaming columns
rename = {
    "med_4___1": "hx_none",
    "med_4___2": "hx_diabetes",
    "med_4___3": "hx_obesity",
    "med_4___4": "hx_heart",
    "med_4___5": "hx_lung",
    "med_4___6": "hx_rheum",
    "med_4___7": "hx_kidney",
    "med_4___8": "hx_liver",
    "med_4___9": "hx_allergies",
    "med_4___10": "hx_pain",
    "med_4___11": "hx_gi",
    "med_4___12": "hx_ent",
    "med_4___13": "hx_neuro",
    "med_4___14": "hx_other",
    
    "med_5___1": "mhx_none",
    "med_5___2": "mhx_mood",
    "med_5___3": "mhx_anxiety",
    "med_5___4": "mhx_psych",
    "med_5___5": "mhx_eating",
    "med_5___6": "mhx_neurodev",
    "med_5___7": "mhx_sleep",
    "med_5___8": "mhx_suicide",
    "med_5___9": "mhx_learning",
    "med_5___10": "mhx_other",
    "med_5___11": "mhx_undisclosed",
    
    "lifestyle1": "alcohol",
    "lifestyle2": "binge",
    "lifestyle3": "smoke",
    "lifestyle4": "exercise"
    
}
df.rename(columns=rename, inplace=True)

df["hx_any"] = df.apply (lambda row: row.hx_diabetes or row.hx_obesity or row.hx_heart or row.hx_lung or row.hx_rheum or row.hx_kidney or row.hx_liver or row.hx_allergies or row.hx_pain or row.hx_gi or row.hx_ent or row.hx_neuro or row.hx_other , axis=1)
df["mhx_any"] =  df.apply (lambda row: row.mhx_mood or row.mhx_anxiety or row.mhx_psych or row.mhx_eating or row.mhx_neurodev or row.mhx_sleep or row.mhx_suicide or row.mhx_learning or row.mhx_other, axis=1)

df
```

```python
df.to_parquet(OUTPUT_PATH+"/survey/"+"once_merged.parquet")
```

# Process all Daily Surveys

```python
daily = ["daily_stress_measure.csv.gz", "cough_ema.csv.gz", "selfassessment_mannequin_sam.csv.gz",
        "ppe_and_exposure_scale.csv.gz"]

#DROPPING DUPLICATES KEEP FIRST
daily_dfs = [pd.read_csv(DATAPATH+EXTENSION+NAME,compression='gzip') for NAME in daily]

#For each df, create a clean date-time column to merge oon
for df in daily_dfs:
    #get appropriate scheduledate column
    date_col = [col for col in df.columns if 'scheduledate' in col]
    #create new column with a clean date-time
    df["date"] = df.apply (lambda row: pd.to_datetime(row[date_col[0]]).date(), axis=1)
    
```

```python
daily_dfs[1]
```

```python
daily_merged = reduce(lambda  left,right: pd.merge(left,right,on=['participant_id', 'date'],
                                            how='outer'), daily_dfs)
```

```python
daily_merged
```

```python
daily_merged = daily_merged[daily_merged['date'].notna()]
```

```python
df = daily_merged.loc[:,~daily_merged.columns.duplicated()]

df.to_parquet(OUTPUT_PATH+"/survey/"+"daily_merged.parquet")
```
